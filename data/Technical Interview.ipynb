### Part 1: Experiment Design

In this part of the challenge I'm going to explain how could I measure the impact of the professional photography service in detail. Explain which would be my experiments and how would I make sure they are not biased.

Firstly, there is obviously a general overview of the generic situation. I would start by doing some basic statistics. Starting with the conversion rate from the professional photos (from now on PP) versus the casual photos (from now on CP), the conversion rate would be calculated by dividing the number of booked houses by the total number of houses, I would calculate this for each day of the past year. At least that is a start. After I would to the same but clustering them by size, city and average booked price. Okey, so now we can know which ones have better conversion rate. The fact of clustering them by size, location and average booked price per day would help me to not have a biased experiment since each an every conversion rates house would be compared between houses with the same traits, the only difference being if they were PP or CP. In this case, we are trying to find out what is best for the booking rate.

I think it is also important to take into consideration the average clients reviews, since having different type of pictures might give a different type of idea to the consumer, maybe one that it is not the real one. I think that in this case we would not only cluster them by the same previous segments but we would also add the number of pictures of the house, since the more picture the more information the user will have about the house. In this case, we are looking to see if the proffesional pictures are making the houses look better than they really are or if on the other hand, they actually give the clients better insights about what they are going to find once they arrive to the house. In this case, we are trying to find out what is best for the customer satisfaction.

Once we start to have an idea about the conversion rate an the clients satisfaction, it is time to know if a customer would actually pay more for a PP than for a CP. I have come up with 2 different ways to validate this hypothesis. Firstly, the one that I think that has the more sense is an A/B testing. Showing to different customers 4 different types of postings (obviously this test should be done with a low number of users since it could make Airbnb look really bad). In order to verify the hypothesis there should be compared the conversion rate of PP with low and high price (low meaning a 5% or 10% cheaper) and the CP with low and high price. The fact of checking with two different prices is to understand the different customer behaviours that there can be.

Another way that I think it would be also be a good way of knowing if the client would be willing to pay more for different type of photos would be by comparing one years average booking price with CP and the next years booking price with PP. In order to validate this experiment, there should be some things taken into account: The house must not have had any type of significant repair (one that it can be seen in the pictures or added in the description). The price difference would also have to be compared with the price difference of the houses around in the same zone. Let's say that with the CP the average booking price was 100 and with the PP 150. Only knowing this price difference and that there has not been any repair would not be enough to be able to say that it makes a difference because it can be caused by an external factor. Therefore we should check if the 50% increase is consistent with the houses with similar traits around or if it's just this house. For example if the average booking price of the other houses was 120 CP and now is 150 PP, it's an start.
It goes without saying that this experiment should be done with thousands of houses.
